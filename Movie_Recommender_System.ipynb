```python
movies.info()

# Remove Columns form 23 columns which does not help in analysis
# Remove -> Budget , homepage , original_language[4510 is english movies] , Popularity , production_companies , original_title , release_date , revenue , runtime , spoken_language, tagline ,status ,vote_count,vote_average,movie_id  
# Important Columns - > genres,id,keywords , title , overview , cast , crew
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# Make a dataframe consisting [movie_id|title|tags]
# Tags consists of columns-> Overview,Genres,keywords,cast[top 3],crew[only director]


movies.isnull().sum()
```

```python
#Drop NAN Valued rows
#inplace=True → Makes the change directly in the original movies
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#Check for duplicated data
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#It is in the form of list of dictionary,we need to convert it into ['Sci-Fi','Action','Fantasy'] format
#Create a helper function
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
#Same like genres do for other columns

movies['keywords'] = movies['keywords'].apply(convert)

```

```python
#Only need top 3 cast
def convertTopThree(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i['name'])
            counter += 1
        else:
            break;
    return L

movies['cast'] = movies['cast'].apply(convertTopThree)
```

```python

#{"credit_id": "52fe48009251416c750aca23", "department": "Editing", "gender": 0, "id": 1721, "job": "Editor", "name": "Stephen E. Rivkin"}
#need to extract name from dictionary where job:"director"

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#Overview is a string convert it to a list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# 'Sam Mendes'->'SamMendes' remove all spaces so apply transformation

movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#tags column by concatinating all other column

movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] 
movies.head()
```

```python
# Create new dataframe ,keep just 3 columns id,title,tags

new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x)) #Convert the list in tags to string format
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#Stemming -> if values are ['activity','activities'] it will be converted to ['activity','activity']

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)


new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
  
```

```python
#By default values are sparse so convert to numpy matrix

vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
# len(cv.get_feature_names_out())
```

```python
#Calculate the distance between the vectors[more distance less similiarity]
#Use cosine distance - angle between the vectors [less angle more similiarity] Euclidean distance is not good measure in highier dimension [here 5000 vectors]

from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
#Total distance calculated is 4806 movies X 4806[each movie with 4806 vectors]
```

```python
similarity[0]  #[Diagonal is always 1]
```

```python
#Once we perform sorting we cannot find which movie distance is calculated with the movie given,hence we use enumeration which will return list of tuples
list(enumerate(similarity[0]))
```

```python
sortedList = sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1]) #sorting need to be done on 2st index not 1th because its 1
sortedList[1:6] #Top 5 similarity to the first movie
```

```python
#Recommend 5 movies from given movie
#Step-1 Fetch the index of the given movie
#Step-2 Search in similiarity matrix with the help of index
#Step-3 Sort the values[less[high similarity] to high value[less similarity]] and return top 5

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x :x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
    
# recommend('Batman Begins')
```
